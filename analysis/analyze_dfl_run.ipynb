```python
import os
import json
```

```python
run_path = "/Users/samacquaviva/Documents/projects/dfl2dsl/experiments_iterative/outputs/dfl_baselines/domains/re2/dfl_120/seed_0/dfl_120_all"
```

```python
n_iterations = int(max([f for f in os.listdir(run_path) if f.isdigit()], key=int)) + 1
```

```python
# Print task->python results.
for i in range(n_iterations):
    python_res_f = os.path.join(run_path, str(i), "train", "gpt_solver_python_results.json")
    with open(python_res_f, "r") as f:
        python_res = json.load(f)
    
    # Map task to (program, correct?).
    task_to_prog = {}
    for parse_info in python_res["parse_results_valid"]:
        task = parse_info["task_attempted"]
        correct = False
        if not task:
            task = parse_info["tasks_solved"][0]
            correct = True
        prog = parse_info["program"]
        task_to_prog.setdefault(task, []).append((prog, correct))

    # Pretty print.
    print(f"Iteration {i}")
    for task, ps in task_to_prog.items():
        print(f"Task: {task}")
        for prog, correct in ps:
            print(prog)
            print(f"Correct: {correct}")
            print()
```

```python
# Print primitives
for i in range(n_iterations):

    grammar_f = os.path.join(run_path, str(i), "laps_grammar.json")
    with open(grammar_f, "r") as f:
        grammar_log = json.load(f)

    prim_names = list(grammar_log["definitions"].keys())
    prim_names.sort()

    print(f"Iteration {i}")
    for prim_name in prim_names:
        print(f"{prim_name}: {grammar_log['definitions'][prim_name]}")
    print()
```
